# DEPOD data preprocessing
1. DEPOD 데이터에서 Substrate Peptide의 길이를 11로 바꿔준다.
2. Kinase와 Peptide를 '_'를 기준으로 이어준다.

```python
depod_ = depod.loc[:,["Phosphorylating protein kinases", "Substrate"]]
depod_.head(3)
depod = depod_
```

```python
depod['Substrate'] = depod['Substrate'].apply(lambda x:x[2:13])
depod['kin_sub'] = depod['Phosphorylating protein kinases'].str.cat(depod['Substrate'], sep='_')
depod.head(3)
```

```python
# depod.to_csv("/home/hb/python/pp_pred/EDA/result_220614_pp-pred_(networKIN&GPS&DEPOD비교)/depod_preprocessing.csv")
```

```python
depod_kinsub = depod.loc[:,['kin_sub']]
depod_kinsub
```

# GPS data prepreocessing
1. GPS 데이터에서 Peptide의 길이를 11로 바꿔준다.
2. Kinase와 Peptide를 '_'를 기준으로 이어준다.

```python
# df_info(gps)

gps['Peptide'] = gps['Peptide'].apply(lambda x:x[2:13])
gps['kin_sub'] = gps['Kinase'].str.cat(gps['Peptide'], sep='_')

gps['kin_sub'] = gps['kin_sub'].apply(lambda x:x.split('/'))
gps['kin_sub'] = gps['kin_sub'].apply(lambda x:x[-1])
gps.head(3)
```

```python
# gps.to_csv("/home/hb/python/pp_pred/EDA/result_220614_pp-pred_(networKIN&GPS&DEPOD비교)/gps5.0_preprocessing.csv")
```

```python
gps_kinsub = gps.loc[:,['kin_sub']]
gps_kinsub
```

# NetworKIN data preprocessing

1. NetworKIN 데이터의 Peptide는 중간에 소문자가 끼어있으므로, 이를 대문자화해준다.
2. Kinase와 Peptide를 '_'를 기준으로 이어준다.

```python
df_info(networkin)
```

```python
networkin['Peptide sequence window'] = networkin['Peptide sequence window'].apply(lambda x:x.upper())
networkin['kin_sub'] = networkin['Kinase/Phosphatase/Phospho-binding domain description'].str.cat(networkin['Peptide sequence window'], sep='_')

networkin.head(3)
```

```python
# networkin.to_csv("/home/hb/python/pp_pred/EDA/result_220614_pp-pred_(networKIN&GPS&DEPOD비교)/networkin_preprocessing.csv")
```

```python
networkin_kinsub = networkin.loc[:,['kin_sub']]
networkin_kinsub
```
